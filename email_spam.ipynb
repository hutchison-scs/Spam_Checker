```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split # Added this
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# 1. Load Data
df = pd.read_csv('emails.csv')

print (df.columns)

```

```python
# 2. Split Data
X = df['text']
y = df['label']
# Correct way:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 3. Create the Model Pipeline
# This converts text to numbers (TF-IDF) then applies the Naive Bayes logic
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
# 4. Train the Model
model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
new_emails = [
    "Hey, are we still on for lunch at the cafe?",
    "URGENT: Your account has been compromised. Click here to reset your password and claim $500!",
    "Discount code inside! Save 20% on your next order."
]

# Run the prediction
results = model.predict(new_emails)

# Display the results
for email, label in zip(new_emails, results):
    print(f"Message: {email}")
    print(f"Prediction: {label}\n")
```
